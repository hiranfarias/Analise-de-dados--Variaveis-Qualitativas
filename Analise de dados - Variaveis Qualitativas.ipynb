Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada? 

    --> Não, não se alterou muito ao longo do tempo

````python

ax = sns.countplot(x='data_ref', hue = 'posse_de_imovel', data=df_previsao_rendas)

df_indicador = df_previsao_rendas['data_ref'].map(lambda x : x).unique()

ax.set_xticks(list(range(df_previsao_rendas['data_ref'].nunique())))

ax.set_xticklabels(df_indicador, rotation=45, ha='right')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0,).set(title='Posse de Imóvel');
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python

def df_grafico(var):

    ax = sns.pointplot(x="data_ref", y="posse_de_veiculo", hue = var ,data = df_previsao_rendas, dodge=True, ci = 95)
    plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
    df_indicador = df_previsao_rendas['data_ref'].map(lambda h : h).unique()
    ax.set_xticks(list(range(df_previsao_rendas['data_ref'].nunique())))
    ax.set_xticklabels(df_indicador, ha='right',rotation=45)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0,).set(title=var);
````
